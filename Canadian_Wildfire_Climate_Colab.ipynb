```python
# ðŸ“‚ Step 2: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## ðŸ”— Step 3: Load the Datasets

**Note**: Replace these URLs with working CSVs from Government of Canada and temperature anomaly datasets.


```python
# Placeholder NBAC data: burned area by year (replace with actual file/URL)
# nbac_url = 'https://raw.githubusercontent.com/adamkashtanov/data/main/nbac_burned_area.csv'
# If the URL is not working, create a placeholder DataFrame
nbac = pd.DataFrame({
	'Year': [2015, 2016, 2017, 2018, 2019, 2020],
	'Burned_Ha': [2500000, 1800000, 3200000, 2100000, 2700000, 1500000]
})
nbac.head()
```

```python
# Placeholder temperature anomaly data (Berkeley Earth or NASA GISTEMP)
temp_url = 'https://raw.githubusercontent.com/datasets/global-temp/master/data/annual.csv'
temp = pd.read_csv(temp_url)
temp = temp.rename(columns={"Mean": "Temp_Anomaly"})[['Year', 'Temp_Anomaly']]
temp.head()
```

```python
# Merge datasets
df = pd.merge(nbac, temp, on='Year', how='inner')
df.head()
```

## ðŸ“ˆ Step 4: Visualizations

```python
# Plot Burned Area by Year
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Burned_Ha', marker='o')
plt.title('Canadian Burned Area per Year (NBAC)')
plt.ylabel('Burned Area (hectares)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
# Scatterplot: Temp Anomaly vs. Burned Area
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='Temp_Anomaly', y='Burned_Ha')
plt.title('Burned Area vs. Temperature Anomaly')
plt.xlabel('Temperature Anomaly (Â°C)')
plt.ylabel('Burned Area (hectares)')
plt.grid(True)
plt.show()
```
